```python
import heapq
current_node = start_node

# Color map for nodes
node_colors = ["gray"] * (rows * cols)
node_colors[start_node] = "green"
node_colors[goal_node] = "red"

# Edge colors for visualization
edge_colors = ['darkgray' for _ in G_grid.edges()]

# Initialize the search frontier with the starting node
frontier = []  # This will replace the 'open' list
heapq.heappush(frontier, (0, current_node))  # Push the start node with priority 0 (or based on heuristic if you prefer)

# Relay nodes to assist in bridging gaps in the search frontier
relay_nodes = []  # This will track potential relay nodes

# Closed list to track fully explored nodes
closed = []

import heapq

# Function to calculate Manhattan distance between two nodes
def manhattan_distance(point1, point2):
    return abs(point2[0] - point1[0]) + abs(point2[1] - point1[1])
```

```python
def update(a):
    global current_node

    # Identify neighbors of the current node
    neighbors = list(G_grid.neighbors(current_node))

    # Explore the neighbors and update the frontier
    for neighbor in neighbors:
        if neighbor not in closed:
            h = manhattan_distance(pos[neighbor], pos[goal_node])
            # Push neighbor into the min heap with priority based on the heuristic h(n)
            heapq.heappush(frontier, (h, neighbor))

    # Mark the current node as closed and update its color
    closed.append(current_node)
    node_colors[current_node] = "pink"

    # Update edge colors
    for edge in G_grid.edges(current_node):
        if edge[0] > edge[1]:
            edge = edge[::-1]
        edge_colors[list(G_grid.edges()).index(edge)] = 'yellow'

    # If the frontier is empty and the goal hasn't been reached, stop the animation
    # print(len(frontier))
    if not frontier:
        print('Cannot reach the goal')
        anim.event_source.stop()
        return

    # Select the next node from the frontier
    _, current_node = heapq.heappop(frontier)
    node_colors[current_node] = "green"

    # Check for nodes with all neighbors closed and revert them to gray

    for node in closed:
        if all(neighbor in closed for neighbor in G_grid.neighbors(node)):
            node_colors[node] = "gray"
            closed.remove(node)  # Remove the node from the closed list as it's no longer fully closed

    # Clear the axes and redraw the updated graph
    ax.clear()
    for _, node in frontier:
        x, y = pos[node]
        circle = plt.Circle((x, y), radius=0.8, color='greenyellow', alpha=0.5)  # Adjust the radius and color as needed
        ax.add_patch(circle)
    nx.draw(G_grid, pos, with_labels=False, node_size=node_size, node_color=node_colors, font_size=font_size, font_weight="bold", edge_color=edge_colors,width=4.2)

    # Overlay the frontier mask

    # Check if the goal has been reached
    if goal_node in neighbors:
        print('Goal found')
        anim.event_source.stop()
```

```python
import random
import numpy as np

# Define parents P1 and P2
parent1 = ['D', 'I', 'J', 'O', 'N', 'H', 'G', 'B', 'C', 'K', 'F', 'L', 'A', 'M', 'E']
parent2 = ['H', 'B', 'F', 'N', 'K', 'J', 'I', 'C', 'G', 'A', 'D', 'M', 'O', 'E', 'L']



# Function to convert characters to integers
def char_to_int(char_list):
    return np.array([ord(char) - ord('A') for char in char_list])

# Convert parents P1 and P2 to integers
parent1_int = char_to_int(parent1)
parent2_int = char_to_int(parent2)

def OX_crossover(parent1, parent2):
    '''
    parent1 and parent2 are 1D np.array
    '''
    cutoff_1, cutoff_2 = 5, 10

    # Initialize offspring
    offspring1 = np.zeros(len(parent1), dtype=parent1.dtype)
    offspring2 = np.zeros(len(parent2), dtype=parent2.dtype)

    # Copy a subset from parent1 to offspring1
    offspring1[cutoff_1:cutoff_2] = parent1[cutoff_1:cutoff_2]

    # Copy a subset from parent2 to offspring2
    offspring2[cutoff_1:cutoff_2] = parent2[cutoff_1:cutoff_2]

    # Fill remaining positions in offspring1 with elements from parent2
    remaining_indices = np.concatenate([np.arange(0, cutoff_1), np.arange(cutoff_2, len(parent1))])
    remaining_values = [val for val in parent2 if val not in offspring1[cutoff_1:cutoff_2]]
    offspring1[remaining_indices] = remaining_values[:len(remaining_indices)]

    # Fill remaining positions in offspring2 with elements from parent1
    remaining_indices = np.concatenate([np.arange(0, cutoff_1), np.arange(cutoff_2, len(parent2))])
    remaining_values = [val for val in parent1 if val not in offspring2[cutoff_1:cutoff_2]]
    offspring2[remaining_indices] = remaining_values[:len(remaining_indices)]

    return offspring1, offspring2

# Perform OX crossover
offspring1, offspring2 = OX_crossover(parent1_int, parent2_int)

# Convert offspring back to characters
def int_to_char(int_list):
    return [chr(i + ord('A')) for i in int_list]

offspring1_char = int_to_char(offspring1)
offspring2_char = int_to_char(offspring2)

# Print the offspring
print("Offspring 1:", ' '.join(offspring1_char))
print("Offspring 2:", ' '.join(offspring2_char))
```
